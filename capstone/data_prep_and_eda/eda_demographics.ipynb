### EDA on Demographics Data
- city: city
- state: state
- median_age: median age of the city
- male_population: number of males in the city
- female_population: number of females in the city
- total_population: total number of population
- num_of_veterans: number of veterans in the city
- foreign_born: (immigrants) citizens who borned in other countries
- avg_house_size: average house size in the city
- state_code: US state code
- race: demographical metadata - race
- count: count
- city_state_code: a combined field, whihc is consist of city and US stat code
